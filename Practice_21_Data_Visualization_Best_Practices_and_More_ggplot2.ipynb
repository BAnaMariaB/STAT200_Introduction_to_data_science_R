####<font color=red>Task 2</font>


Gross domestic product (GDP) is an economic indicator quantifying the size of an economy. Countries with higher GDP typically have stronger economies. Let's analyze this metric.

<font color=red>Task 2.1</font>

* Use `ggplot()` to plot a histogram of `gdp` for the `gapminder3` data frame.
  * Be sure to include a title and axes labels using the `labs()` layer

* What are your observations about the shape of the distribution of `gdp`?

```
#2.1
ggplot(gapminder3, aes(x = gdp)) +
  geom_histogram(bins = 30) +
  labs(
    title = "GDP (1970, 1990, 2010)",
    x = "GDP",
    y = "Count"
  )
```

**Task 2.1**<br>
* Distribution of GDP is right-skewed.  
* The histogram shows a large concentration of observations near zero, so most countries have low GDP values.
* A long tail toward the right, with a small number of countries having extremely high GDP.


<br>

<font color=red>Task 2.2</font>

* Apply a log transformation to the `gdp` variable that will make the data easier to visualize.

* Store the transformed `gdp` variable as another variable `log_gdp` in the `gapminder3` data frame

* Plot a histogram of `log_gdp`

```
#2.2
gapminder3$log_gdp <- log(gapminder3$gdp)

ggplot(gapminder3, aes(x = log_gdp)) +
  geom_histogram() +
  labs(
    title = "Histogram of log(GDP)",
    x = "log(GDP)",
    y = "Count"
  )
```

<br>

<font color=red>Task 2.3</font>

* Plot the histogram from Task 2.2 with its continuous density function (density should now be on the y-axis)

* What are your observations about the distribution of log GDP?
  * Is the distribution skewed?
  * Is the distribution multimodal?
  * etc.

```
#2.3
ggplot(gapminder3, aes(x = log_gdp)) +
  geom_histogram(aes(y = ..density..)) +
  geom_density(alpha = 0.2) +
  labs(
    title = "Histogram and Density of log(GDP)",
    x = "log(GDP)",
    y = "Density"
  )
```

**Task 2.3**<br>
Distribution of log(GDP) appears roughly bell-shaped and approximately symmetric, with slight right skewness and it's unimodal.

<br>

<font color=red>Task 2.4</font>

* Plot the histogram and density from Task 2.4 but also include the `year` variable in a `facet_wrap()` layer

  * Be sure to specify `nrow = 3` in the `facet_wrap()` layer

* What are your observations of the distribution of log GDP over the three years?

```
#2.4
ggplot(gapminder3, aes(x = log_gdp)) +
  geom_histogram(aes(y = ..density..)) +
  geom_density(alpha = 0.2) +
  facet_wrap(~ year, nrow = 3) +
  labs(
    title = "Histogram and Density of log(GDP) over 3 periodes of Time",
    x = "log(GDP)",
    y = "Density"
  )
```

**Task 2.4** <br>
* Distribution of log(GDP) is unimodal and approximately symmetric in 1970, 1990, and 2010, with slight right skewness across all years.
* Over the years, the center of the distribution has shifted slightly to the right.
* This means an overall increase in GDP across countries.

####<font color=red>Task 3</font>


<font color=red>Task 3.1</font>

* Use `ggplot()` to plot a boxplot with `log_gdp` on the y-axis and `region` on the x-axis from the `gapminder3` data frame

  * You can use the script below to resize your boxplot

    `options(repr.plot.width = 25, repr.plot.height = 8)`

* Label your axes appropriately

```
#3.1
options(repr.plot.width = 25, repr.plot.height = 8)

ggplot(gapminder3, aes(x = region, y = log_gdp)) +
  geom_boxplot() +
  labs(
    title = "Boxplot of log(GDP) by Region",
    x = "Region",
    y = "log(GDP)"
  )
```

<br>

<font color=red>Task 3.2</font>

* Order the categories in your boxplot from smallest median log GDP to largest median log GDP

    `options(repr.plot.width = 25, repr.plot.height = 8)`

* What are your observations?
  * Which region has the lowest/highest log GDP?
  * Which region has the highest variability in log GDP?
  * Compare some regions
  * etc.

```
#3.2
gapminder3$region <- reorder(gapminder3$region, gapminder3$log_gdp, FUN = median)
options(repr.plot.width = 25, repr.plot.height = 8)

ggplot(gapminder3, aes(x = region, y = log_gdp)) +
  geom_boxplot() +
  labs(
    title = "Boxplot of log(GDP) by Region (ordered by median)",
    x = "Region",
    y = "log(GDP)"
  ) +
  coord_flip()
```

**Task 3.2** <br>
* Northern America has the highest median log(GDP)
* Micronesia has the lowest median log(GDP)
<br>
<br>
* Eastern Asia has the highest variability in log(GDP), this hits at a mix of rich and poor countries in the region.
<br>
<br>
* Northern America and Western Europe have high GDPs, low variability
* African regions have lower GDP levels and large variability
* Micronesia and Polynesia have consistently low GDPs, little spread.

<br>

<font color=red>Task 3.3</font>

* Plot your boxplot from Task 3.3 with a jitter geometry layer

  * Use the appropriate `width` and `alpha` arguments to make the visualization easier to interpret

* Which regions have very few data points?

```
#3.3
ggplot(gapminder3, aes(x = region, y = log_gdp)) +
  geom_boxplot(outlier.shape = NA) +
  geom_jitter(width = 0.2, alpha = 0.6) +
  labs(
    title = "Boxplot with Jittered Points of log(GDP) by Region",
    x = "Region",
    y = "log(GDP)"
  ) +
  coord_flip()
```

**Task 3.3** <br>
Micronesia, Polynesia have very few data points

<br>

<font color=red>Task 3.4</font>

* Include `year` as a fill argument to the boxplot from Task 3.2 (no jitter)

  Hint: `year` must be a factor in this case


* What are your observations of log GDP in each region across the three years?

```
#3.4
gapminder3$year <- as.factor(gapminder3$year)

ggplot(gapminder3, aes(x = region, y = log_gdp, fill = year)) +
  geom_boxplot() +
  labs(
    title = "Boxplot of log(GDP) by Region and Year",
    x = "Region",
    y = "log(GDP)"
  ) +
  coord_flip()
```

**Task 3.4** <br>
* All regions, the log(GDP) increased from 1970 to 1990 to 2010 meaning global economic growth.
* The growth was stronger in regions such as Northern America, Eastern Asia, and Western Europe,
* Melanesia showed minimal change.

<br>

<font color=red>Task 3.5</font>

* Change the colors of `year` in your boxplot from `Task 3.3` to be colorblind friendly using a `scale_fill_brewer()` layer

* Use the `"Dark2"` palette
    

```
#3.5

ggplot(gapminder3, aes(x = region, y = log_gdp, fill = year)) +
  geom_boxplot() +
  scale_fill_brewer(palette = "Dark2") +
  labs(
    title = "Boxplot of log(GDP) by Region and Year (Colorblind Friendly)",
    x = "Region",
    y = "log(GDP)"
  ) +
  coord_flip()
```

<font color=red>Task 4.1</font>

* Use `ggplot()` to plot a scatter plot with `life_expectancy` on the y-axis and `log_gdp` on the x-axis from the `gapminder3` data frame

```
#4.1
ggplot(gapminder3, aes(x = log_gdp, y = life_expectancy)) +
  geom_point() +
  labs(
    title = "Life Expectancy vs log(GDP)",
    x = "log(GDP)",
    y = "Life Expectancy"
  )
```

**Task 4.1** <br>
There's positive relationship between log(GDP) and life expectancy: as log(GDP) increases, life expectancy tends to rise.

<br>

<font color=red>Task 4.2</font>

* Plot your scatter plot from Task 4.1 with a color visual dimension for `continent`

* Also, increase the overall size of the points by including a `size = ` argument in the point geometry layer

```
#4.2
ggplot(gapminder3, aes(x = log_gdp, y = life_expectancy, color = continent)) +
  geom_point(size = 5) +
  labs(
    title = "Life Expectancy vs log(GDP) (by continent)",
    x = "log(GDP)",
    y = "Life Expectancy"
  )
```

<br>

* Include a `facet_wrap` layer for `year` to your plot from Task 4.2

  * etc.

```
#4.3
ggplot(gapminder3, aes(x = log_gdp, y = life_expectancy, color = continent)) +
  geom_point(size = 2) +
  facet_wrap(~ year) +
  labs(
    title = "Life Expectancy vs log(GDP) (by continent and year)",
    x = "log(GDP)",
    y = "Life Expectancy"
  )
```
