<font color=red>Task 5.2</font>

* Plot a heatmap with `year` on the y-axis and `life_expectancy_highlow` on the x-axis using `ggplot()`.

* "Add" the layer below to change the color of your plot

  `scale_fill_gradient(low = "white", high = "red")`

* Include labels for the title, x-axis, and y-axis

* What trend do you observe over time? Can you explain the trend you observe?

```
#5.2
heatmap_data <- as.data.frame(table(gapminder$year, gapminder$life_expectancy_highlow))
names(heatmap_data) <- c("year", "life_expectancy_highlow", "count")

ggplot(heatmap_data, aes(x = life_expectancy_highlow, y = as.factor(year), fill = count)) +
  geom_tile() +
  scale_fill_gradient(low = "white", high = "magenta") +
  labs(
    title = "Life Expectancy over Time",
    x = "Life Expectancy category",
    y = "Year"
  )
```

####<font color=red>Task 6</font>

<font color=red>Task 6.1</font>

* Subset the gapminder dataset to only include the `year` 2000 and name this new data frame `gapminder2000`

* Create a scatterplot with life expectancy on the y-axis and fertility on the x-axis using `ggplot()`

* Include labels for the title, x-axis, and y-axis

* What are your observations?

```
#6.1
gapminder2000 <- gapminder[gapminder$year == 2000, ]

ggplot(gapminder2000, aes(x = fertility, y = life_expectancy)) +
      geom_point(color="magenta") +
      labs(
        title = "Life Expectancy vs Fertility - Year 2000",
        x = "Fertility (birth per woman)",
        y = "Life Expectancy (year)"
      )
```

**Task 6.1**<br>
* A higher number of births per woman is associated with lower life expectancy.

<br>

<font color=red>Task 6.2</font>

* Include `continent` as a shape visual dimension to your plot from Task 7.1

```
#6.2
ggplot(gapminder2000, aes(x = fertility, y = life_expectancy, shape = continent)) +
      geom_point(color="magenta") +
      labs(
        title = "Life Expectancy vs Fertility - Year 2000",
        x = "Fertility (birth per woman)",
        y = "Life Expectancy (year)"
      )
```

<br>

<font color=red>Task 6.3</font>

* Choose two other variables and include as a color and size visual dimension in your plot from Task 6.2

* What are your observations?

```
#6.3
ggplot(gapminder2000, aes(x = fertility, y = life_expectancy, shape = continent,
                          color = gdp, size = population)) +
      geom_point(color="magenta") +
      labs(
        title = "Life Expectancy vs Fertility - Year 2000",
        x = "Fertility (birth per woman)",
        y = "Life Expectancy (year)",
        color = "GDP",
        size = "Population"
      )
```

**Task6.3**<br>
Countries with higher fertility show lower life expectancy. Generally the largest population size vary in areas with moderate fertility rates
